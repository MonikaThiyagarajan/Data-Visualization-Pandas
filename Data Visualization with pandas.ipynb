# Data Visualization

##### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Iris.csv")
```

```python
data.head(2)
```

```python
a = sns.PairGrid(data)
a.map_diag(sns.distplot)
a.map_upper(sns.kdeplot)
a.map_lower(sns.boxplot)
```

### Facet Grid

```python
df = pd.read_csv("tips.csv")
```

```python
df.head()
```

```python
b = sns.FacetGrid(data = df, col = 'smoker')
```

```python
b = sns.FacetGrid(data = df, col = 'smoker' , row = 'time')
```

```python
b = sns.FacetGrid(data = df, col = 'smoker' , row = 'time')
b.map(sns.distplot,'total_bill')
```

```python
b = sns.FacetGrid(data = df, col = 'smoker' , row = 'time')
b.map(sns.distplot,'total_bill')
sns.set_style('darkgrid')
```

### Regression Plot

```python
df.head()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df,hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df,col = 'sex')
```
